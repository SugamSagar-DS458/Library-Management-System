## Library Management System

```python
# Initial library data with availability and tracking
library = {
    "Python Basics": {"author": "Test", "available": False, "issued_to": "Sugam"},
    "Data Science": {"author": "Test1", "available": True, "issued_to": None},
    "Harry Potter": {"author": "J.K. Rowling", "available": True, "issued_to": None},
    "Malgudi Days": {"author": "R.K. Narayan", "available": True, "issued_to": None},
    "Gitanjali ": {"author": "Rabindranath Tagore", "available": True, "issued_to": None},
    "Romeo and Juliet": {"author": "William Shakespeare", "available": False, "issued_to": "Sagar"},
    "Five Point Someone: What Not to Do at IIT": {"author": "Chetan Bhagat", "available": True, "issued_to": None},
    "An Era of Darkness": {"author": "Shashi Tharoor", "available": True, "issued_to": None},
    "Midnight's Children": {"author": "Salman Rushdie", "available": True, "issued_to": None},
    "The White Tiger": {"author": "Aravind Adiga", "available": True, "issued_to": None}
}

# Predefined login credentials
credentials = {"Sugam": "Sugam@6632260", "Arihant": "Arihant123" , "Sagar": "Sagar123"}

# Display All Books
def display_books():
    print("\n--- Current Library Books ---")
    if not library:
        print("The library is currently empty.")
        return
    
    for title, details in library.items():
        status = "Available" if details["available"] else f"Issued to {details['issued_to']}"
        print(f"Title: {title} | Author: {details['author']} | Status: {status}")

# Adding a New Book
def add_book():
    print("\n--- Add a New Book ---")
    title = input("Enter book title: ").strip()
    author = input("Enter book author: ").strip()
    
    if title in library:
        print(f"Error: '{title}' already exists in the library.")
    else:
        library[title] = {"author": author, "available": True, "issued_to": None}
        print(f"Success: '{title}' has been added successfully!")

# Issuing a Book
def issue_book():
    print("\n--- Issue a Book ---")
    title = input("Enter the title of the book to issue: ").strip()
    
    if title in library:
        if library[title]["available"]:
            borrower = input("Enter the name of the person issuing the book: ").strip()
            library[title]["available"] = False
            library[title]["issued_to"] = borrower
            print(f"Success: '{title}' has been successfully issued to {borrower}.")
        else:
            print(f"Error: '{title}' is already issued to {library[title]['issued_to']}.")
    else:
        print("Error: Book not found in the library.")

# Returning a Book
def return_book():
    print("\n--- Return a Book ---")
    title = input("Enter the title of the book to return: ").strip()
    
    if title in library:
        if not library[title]["available"]:
            library[title]["available"] = True
            library[title]["issued_to"] = None
            print(f"Success: '{title}' has been successfully returned to the library.")
        else:
            print(f"Error: '{title}' was not issued.")
    else:
        print("Error: Book not found in the library.")

# View Issued Books
def view_issued_books():
    print("\n--- Currently Issued Books ---")
    issued_count = 0
    
    for title, details in library.items():
        if not details["available"]:
            print(f"Title: {title} | Issued To: {details['issued_to']}")
            issued_count += 1
            
    if issued_count == 0:
        print("No books are currently issued.")

# Login System (Uses recursion instead of a while loop)
def login(attempts=3):
    if attempts == 3:
        print("Welcome to the Library Management System")
        print("----------------------------------------")
        
    if attempts == 0:
        print("\nToo many failed attempts. Exiting system.")
        return False
        
    username = input("Enter Username: ").strip()
    password = input("Enter Password: ").strip()
    
    if username in credentials and credentials[username] == password:
        print(f"\nLogin successful! Welcome, {username}.\n")
        return True
    else:
        print(f"Invalid credentials. Attempts remaining: {attempts - 1}")
        return login(attempts - 1)

# Main Menu (Uses recursion instead of a while loop)
def main_menu():
    print("\n===== MAIN MENU =====")
    print("1. View Books")
    print("2. Add a Book")
    print("3. Issue a Book")
    print("4. Return a Book")
    print("5. View Issued Books")
    print("6. Exit")
    
    choice = input("Enter your choice (1-6): ").strip()
    
    if choice == "1":
        display_books()
    elif choice == "2":
        add_book()
    elif choice == "3":
        issue_book()
    elif choice == "4":
        return_book()
    elif choice == "5":
        view_issued_books()
    elif choice == "6":
        print("\nThank you for using the Library Management System. Goodbye!")
        return  # Breaks the recursion and exits
    else:
        print("Invalid choice! Please enter a number between 1 and 6.")
        
    # Call the menu again to keep it running
    main_menu()

# Program Start Execution
if __name__ == "__main__":
    if login():
        main_menu()
```
